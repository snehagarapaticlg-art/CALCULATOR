```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 150, 120, 180, 200]
plt.plot(months, sales)
plt.title
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'SQL']
students = [40, 30, 20, 10]

plt.pie(students,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Students Learning Programming Languages")

plt.show()
```
